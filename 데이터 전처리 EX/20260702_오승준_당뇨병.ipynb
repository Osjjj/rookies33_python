당뇨병 특성(BMI) 추출 및 데이터 분할 모델 훈련

```python
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression       # 학습모델 import
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score   # 평가지표 함수
from sklearn.model_selection import train_test_split    # 데이터 분리 => 8:2

from sklearn import set_config

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
set_config(display='text')
```

TODO : Mission 1

```python
# diabetes 데이터 로드
diabetes = load_diabetes()

# 2차원 데이터 구조를 유지하기 위한 데이터 프레임 변환
df = pd.DataFrame(diabetes.data)

# 특성(X_bmi)과 타겟(y_bmi) 설정
X_bmi = df[[2]]
y_bmi = diabetes.target

# 훈련 데이터와 테스트 데이터 분리
X_train_b, X_test_b, y_train_b, y_test_b = train_test_split(
    X_bmi, y_bmi,
    test_size = 0.2,
    random_state = 42
)

# 선형회귀 모델 생성
model = LinearRegression()

# 훈련
model.fit(X_train_b, y_train_b)
```

TODO 2 : Mission 2

```python
# 테스트 데이터를 사용하여 예측값 계산
y_pred_b = model.predict(X_test_b)

# 평가지표
mse = mean_squared_error(y_test_b, y_pred_b)
mae = mean_absolute_error(y_test_b, y_pred_b)
rmse = np.sqrt(mse)
r2 = r2_score(y_test_b, y_pred_b)

# 평가지표 출력
print(f'mse : {mse}')
print(f'mae : {mae}')
print(f'rmse : {rmse}')
print(f'r2 : {r2}')
```

TODO : Mission 3

```python
# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 산점도와 회귀선 출력 ( legend로 범례 출력 )
plt.scatter(X_test_b, y_test_b, color ='black', label='실제 데이터')
plt.plot(X_test_b, y_pred_b, color = 'red', label = '예측값 회귀선')
plt.legend()
```
