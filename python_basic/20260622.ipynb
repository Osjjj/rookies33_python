```python
# 파이썬의 데이터 (숫자, 문자열, 논리값) => 객체 형식으로 저장!
# 변수가 가르키는 것을 메모리 주소값
# 객체 : 변수, 함수들
# 변수.함수()
num = 10 # int 정수
num1 = 3.14 # float 실수
str_10 = '10' # str 문자열
```

```python
# 문자열 생성 => ' '  " "  ''' '''  """ """
str_1 = 'hello~'
str_2 = "안녕하세요"
str_3 = '''
첫번째 라인
2
3
네번째 라인
'''
str_4 ="""
select *
from emp
where deptno=10
"""
```
