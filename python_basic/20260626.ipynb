##### 클래스의 상속

```python
# 사칙연산을 하는 클래스
# => 클래스 이름, 변수, 함수(메서드), 초기화 함수(생성자)
# 클래스의 이름 대문자로 시작 => FourCal
# 캐멀케이스
# 변수 : 두 개
# 함수(메서드) : 설정, +, *, -, /
class FourCal :
    # 클래스의 멤버 : 변수, 메서드
    
    # 초기화 함수 => 변수 초기화
    def __init__(self, first, second) :
        # 인스턴스 변수 생성과 데이터 할당
        self.first = first
        self.second = second
    
    # 메서드를 통해 변수(인스턴스 변수) 를 생성 => 클래스를 통해서 객체로 만들면 변수가
    # 함수 생성 시에 self를 쓰는 이유 => 인스턴스 멤버를 사용하기 위함!
    def setData(self, first, second):
        # 인스턴스 변수 생성과 데이터 할당
        self.first = first
        self.second = second

    def printData(self):
        print(f'first : {self.first}, second : {self.second}')

    def add(self):
        return self.first + self.second
    
    def sub(self) :
        return self.first - self.second
    
    def mul(self) :
        return self.first * self.second
    
    def div(self) :
        try :
            return self.first / self.second
        except ZeroDivisionError :
            print('0으로 나눠줄 수 없습니다.')
```

```python
# 클래스의 상속 : 새로운 클래스를 만드는 것
# 상속 => 기존 클래스의 멤버들을 받아서 내 것으로 사용
# 오버라이딩 => 메서드를 수정해서 사용, 처리블록을 수정해서 사용
# MoreFourCal : 새로운 클래스 정의 => pow 제곱을 구해주는 함수를 추가, div 메서드 수정
class MoreFourCal(FourCal) :
    # pass
    def pow(self):
        return self.first**self.second
    
    # 오버라이딩 : 함수의 이름과 매개변수는 그대로 유지하고 -> 처리 부분만 수정해서 정의
    def div(self):
        if self.second == 0 :
            return 0
        return self.first/self.second
```

```python
# cal1 = MoreFourCal()  # 에러!!! => 생성자 호출
cal1 = MoreFourCal(3, 0)
```
