### 프로그램의 흐름제어

```python
# if : 분기 => ~라면 , 조건에 따라 처리 방식을 정의하는 방법
# ~라면 => 조건식
# 조건식을 잘 만드는 것이 중요!!!!!

# 성인 판단 => 성인이면 입장, 미성년이면 입장 거부

age = int(input('나이를 입력하세요!'))   # age > 20

# 성인 : 환영 인사, 미성년자 : 끄지
if age >= 20 :
    print('반갑습니다 ~_~')
    print('넷플릭스 영화를 시작하겠습니다. 두둥')
    # play()
else:
    print('끄지십시오~')
```

```python
# 짝수, 홀수 판단
# 짝수와 홀수 판단 => 조건식 => 짝수인가? 홀수인가?
# num%2 == 0 => 짝수, num%2 == 1 => 홀수
# else               else
# 참 => print, 거짓 => print

num = int(input('정수를 입력하세요!'))

if num%2 == 1 :
    # 홀수 : True
    print('홀수!')
else:
    # 홀수 : False => 짝수
    print('짝수!')
```

```python
# if 중첩
# 양수, 음수, 0
# 양수 : num > 0
# 음수 : num < 0
# 0 : num > 0 아니고, num < 0 아닌 숫자

num = 0

if num > 0 :
    # True
    print('양수')
else :
    # 0보다 작거나 같은 숫자
    if num <0 :
        print('음수')
    else :
        print('0')
```

```python
# if~ elif ~ elif~ else
if num > 0 :
    print('양수')
elif num < 0 :
    print('음수')
else :
    print(0)
```
