### 연습문제 풀이 1

```python
kor_sc = 86
eng_sc = 90
math_sc = 88

total_sc = kor_sc + eng_sc + math_sc
average = total_sc / 3

print(f'국어 점수 : {kor_sc}')
print(f'영어 점수 : {eng_sc}')
print(f'수어 점수 : {math_sc}')
print("="*20)
print(f'총점 : {total_sc}')
print(f'평균 : {average}')
```
