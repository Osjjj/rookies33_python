##### 01 숫자를 입력 받아 짝수, 홀수를 판별하는 프로그램

#### **if 문 코딩해보기**

```python
num = int(input('정수를 입력하세요.'))

if num % 2 == 1 :
    print('입력하신 정수는 홀수입니다.')
else :
    print('입력하신 정수는 짝수입니다.')
```

##### 02 숫자를 입력 받아 양수, 음수, 0을 판별하는 프로그램

```python
num = float(input('숫자를 입력하세요.'))

if num > 0 :
    print('입력하신 숫자는 양수입니다.')
elif num < 0 :
    print('입력하신 숫자는 음수입니다.')
else :
    print('입력하신 숫자는 0입니다.')
```

##### 03 독감 예방접종이 가능한 여부를 확인하는 프로그램

```python
age = int(input('나이를 입력해주세요.'))

if age < 15 or age >= 65 :
    print('무료 예방접종 가능!')
else :
    print('무료 예방접종 대상 아님!')
```
