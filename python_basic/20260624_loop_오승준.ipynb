##### 01 1부터 100까지의 합을 구하는 프로그램 작성

```python
num = 1
total_w = 0

# while 문을 이용한 합 계산
while num :
    total_w +=num
    num += 1
    if num > 100 :
        break

print(total_w)

total_f = 0

# for 문을 이용한 합 계산
for i in range(1,101) :
    total_f += i

print(total_f)
```

##### 02 for 문을 이용하여 1부터 10까지를 곱한 결과를 출력하는 프로그램 작성

```python
result = 1

for i in range(1,11) :
    result *= i

print(result)
```

##### 03 구구단의 짝수 단만 출력하는 프로그램 작성

```python
for i in range(1, 10) :
    if i % 2 == 0 :
        print(i, '단')
        
        for j in range(1, i+1) :
            print(f'{i} X {j} = ', i*j)
```

```python
# 강사님 코드 (Step 2 사용, if X)
for i in range(2, 9, 2) :
    print(i, '단')
    for j in range(1, i+1) :
        print(f'{i} X {j} = ', i*j)
```

##### 04 while문을 무한 루프로 구성하여 작성하는 예제

```python
num = 1
sum = 0

while True :
    if num % 2 == 0 or num % 3 == 0 :
        sum += num
        
    if sum > 1000 :
        print(f'현재 값 : {num}, 총합 : {sum}')
        break

    num += 1 
```
