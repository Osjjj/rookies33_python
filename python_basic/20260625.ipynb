```python
# 사칙연산을 하는 클래스
# => 클래스 이름, 변수, 함수(메서드), 초기화 함수(생성자)
# 클래스의 이름 대문자로 시작 => FourCal
# 캐멀케이스
# 변수 : 두 개
# 함수(메서드) : 설정, +, *, -, /
class FourCal :
    # 클래스의 멤버 : 변수, 메서드
    
    # 초기화 함수 => 변수 초기화
    def __init__(self, first, second) :
        # 인스턴스 변수 생성과 데이터 할당
        self.first = first
        self.second = second
    
    # 메서드를 통해 변수(인스턴스 변수) 를 생성 => 클래스를 통해서 객체로 만들면 변수가
    # 함수 생성 시에 self를 쓰는 이유 => 인스턴스 멤버를 사용하기 위함!
    def setData(self, first, second):
        # 인스턴스 변수 생성과 데이터 할당
        self.first = first
        self.second = second

    def printData(self):
        print(f'first : {self.first}, second : {self.second}')

    def add(self):
        return self.first + self.second
    
    def sub(self) :
        return self.first - self.second
    
    def mul(self) :
        return self.first * self.second
    
    def div(self) :
        try :
            return self.first / self.second
        except ZeroDivisionError :
            print('0으로 나눠줄 수 없습니다.')
```

```python
# 객체 생성
# cal1 = FourCal()
cal1 = FourCal(10, 0)
```

```python
print(id(cal1), type(cal1))
```

```python
# 참조 방법
# 변수.인스턴스 변수 이름
# 변수.메서드 이름()
# print(cal1.first)
cal1.setData(10, 7)     # setData(2145378036832, 10, 7))
```

```python
print(cal1.first, cal1.second)
cal1.printData()
print(cal1.add())
print(cal1.sub())
print(cal1.mul())
print(cal1.div())
```
