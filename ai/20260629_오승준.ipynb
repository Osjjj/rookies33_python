##### 머시기미서기임

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('식품_휴게음식점.csv', encoding='cp949', low_memory=False)
```

```python
df_1 = df[['영업상태명', '사업장명', '지번주소']]
df_1
```

```python
df_2 = df_1.dropna(subset=['지번주소'])
df_2
```

```python
df_3 = df_2[ df_2['지번주소'].str.contains('서울특별시')]
```

```python
df_3
```

```python
df_cu = df_3[ df_3['사업장명'].str.contains('씨유|cu', case = False, na = False)]
df_cu
```

```python
df_cu_ing = df_cu[ df_cu['영업상태명'] != '폐업']
df_cu_ing
```

```python
df_gs = df_3[ df_3['사업장명'].str.contains('지에스|gs', case = False, na =False)]
df_gs
```

```python
df_gs_ing = df_gs[ df_gs['영업상태명'] != '폐업']
df_gs_ing
```

```python
df_seven = df_3[ df_3['사업장명'].str.contains('세븐일레븐', na=False)]
df_seven
```

```python
df_seven_ing = df_seven[ df_seven['영업상태명'] != '폐업']
df_seven_ing
```

```python
# 각 편의점 브랜드 별 데이터 프레임을 하나로 통합!

df_convenient = pd.concat([
    df_cu_ing.assign(brand = 'CU'),
    df_gs_ing.assign(brand = 'GS'),
    df_seven_ing.assign(brand = '7ELEVEN')
])


df_convenient
```

```python
# brand 로 그룹화하여 각각의 브랜드 개수 추출
result = df_convenient.groupby('brand').size()
```

```python
# 한글 폰트 설정
plt.rc('font', family = 'Malgun Gothic')

# - 부호 깨짐 방지
plt.rc('axes', unicode_minus = False)
```

```python
plt.bar(result.index, result.values, edgecolor = 'black')
```
