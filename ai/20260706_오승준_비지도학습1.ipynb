```python
from sklearn.ensemble import IsolationForest
import numpy as np
import pandas as pd
```

```python
# 미션용 가상 센서 데이터 생성 (정상 상태 300개 + 오작동 상태 25개 합성)
rng = np.random.RandomState(100)
X_sensor_normal = 0.5 * rng.randn(300, 5)
X_sensor_faulty = rng.uniform(low=-5, high=5, size=(25, 5))
X_sensor = np.vstack([X_sensor_normal, X_sensor_faulty])

print(X_sensor)
sensor_df = pd.DataFrame(X_sensor)
```

```python
sensor_df
```

```python
# 모델 생성
model = IsolationForest(
    n_estimators=100,
    contamination=0.08,
    random_state=42
)
```

```python
outlier_preds = model.fit_predict(sensor_df)
```

```python
outlier_preds
```

```python
# numpy의 sum을 통해 이상치의 개수 계산
anomaly_count = np.sum(outlier_preds == -1)
print(f'이상치 샘플의 개수 : {anomaly_count}개')
```
