##### BeautifulSoup 을 활용한 웹 스크래핑

Mission 1 : books_info

```python
from bs4 import BeautifulSoup
```

```python
html_books = """
<div class='library'>
    <div class='book' id='b-101'>
        <h3 class='book-title'>딥러닝 텐서플로 마스터</h3>
        <p class='book-author'>김작가</p>
    </div>
    <div class='book' id='b-102'>
        <h3 class='book-title'>이것이 진짜 R 프로그래밍</h3>
        <p class='book-author'>박통계</p>
    </div>
    <div class='book' id='b-103'>
        <h3 class='book-title'>자연어 처리 입문</h3>
        <p class='book-author'>최이과</p>
    </div>
</div>
"""
```

```python
# BeautifulSoup 객체 생성!
soup = BeautifulSoup(html_books, 'html.parser')
```

```python
# books_info 리스트 생성
books_info = []

# class가 book인 요소들을 찾아 books 에 저장
books = soup.select('.book')

# 각각의 book 들에서 title과, author를 뽑아내 리스트에 딕셔너리 형태로 추가
for book in books :
    title = book.select_one('.book-title').text
    author = book.select_one('.book-author').text

    books_info.append({
        'title' : title,
        'author' : author
    })

print("=== 추출된 도서 목록 ===")
print(books_info)
```

Mission 2 : news_list

```python
html_news = """
<ul class='news-section'>
    <li class='news-item'><a href='https://news.example.com/ai-1'>인공지능 규제법 통과, IT 업계 파장</a></li>
    <li class='news-item'><a href='https://news.example.com/dev-2'>파이썬 3.12 릴리즈, 무엇이 달라졌나?</a></li>
    <li class='news-item'><a href='https://news.example.com/sec-3'>오픈소스 라이브러리 보안 취약점 경고</a></li>
</ul>
"""
```

```python
soup = BeautifulSoup(html_news, 'html.parser')
```

```python
# news_list 리스트 생성
news_list = []

# 태그가 a를 모두 찾아 news_items 에 저장
news_items = soup.find_all('a')

# 각각의 news 들에서 headline과, url를 뽑아내 리스트에 딕셔너리 형태로 추가
for news in news_items :
    headline = news.text
    
    news_list.append({
        'headline' : headline,
        'url' : news['href']
    })

print("=== 수집된 뉴스 헤드라인 ===")
print(news_list)
```

##### SQLite를 활용한 간단한 데이터베이스 관리

Mission 1 : SQLite 데이터 조작 (CRUD)

```python
import sqlite3
```

```python
# CREATE 문 생성
sql_create = '''
CREATE TABLE IF NOT EXISTS users (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT,
    email TEXT,
    level TEXT
)
'''
```

```python
conn = sqlite3.connect(':memory:')
print('conn..... OKKKKK!!!')

cursor = conn.cursor()

# create문 실행
cursor.execute(sql_create)
conn.close()
```

```python
conn = sqlite3.connect(':memory:')
print('conn..... OKKKKK!!!')

cursor = conn.cursor()

# create문 실행
cursor.execute(sql_create)

# insert문 작성 (C)
sql_insert = 'insert into users (name, email, level) values (?,?,?)'

# 행단위 검색! select 문 생성 ! (R)
# name과 email 컬럼만 조회 where (지정한 level에 맞는)
sql_select = 'select name, email from users where level == ?'

# 데이터 생성
data = [
    ('홍길동', 'hong@example.conm', 'gold'),
    ('이영희', 'lee@example.com', 'silver'),
    ('김철수', 'kim@example.com', 'gold')
]

# conn = sqlite3.connect(':memory:')

# cursor = conn.cursor()

# sql 실행, executemany를 통해 여러 개의 데이터 삽입
cursor.executemany(sql_insert, data)

# 작업 완료! /입력, 수정, 삭제 (CUD) 시 commit 필요
conn.commit()

# select문 실행, level이 gold인 것만 조회함 ( 튜플은 1개일 때 (?,) 것 주의)
cursor.execute(sql_select, ('gold',))

# 모든 조회 결과를 가져와 gold_users 변수에 저장
gold_users = cursor.fetchall()

for user in gold_users :
    print(user)

conn.close()
conn.close()
```

Mission 2 : 수집된 데이터프레임의 DB 테이블 저장 및 복원

```python
import pandas as pd
```

```python
conn = sqlite3.connect('user.db')

cursor = conn.cursor()

# 실습 데이터프레임 생성
df_members = pd.DataFrame({
'member_id': ['M01', 'M02', 'M03'],
'name': ['이민수', '김영희', '박준형'],
'score': [85, 95, 78]
})
```

```python
# to_sql을 통해 df_members 데이터 프레임을 members 라는 이름의 테이블로 DB에 적재!
df_members.to_sql('members', conn, index=False, if_exists='replace')

conn.close()
```

```python
# users.db에 재연결!
conn = sqlite3.connect('users.db')

# members 테이블의 전체 내용을 조회하는 select 문 생성
sql_select = 'select * from members'

# df_read 변수 선언
df_read = None

# df_read 변수에 read_sql을 통해 조회한 전체 내용을 로드함
df_read = pd.read_sql(sql_select, conn)

# df_read 복원 완성!
print("=== DB 복원 데이터프레임 ===")
print(df_read)
```
