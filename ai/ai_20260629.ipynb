```python
# Pandas 를 이용 => 평균, 중간, 최빈
import pandas as pd
```

```python
data = [10, 15, 10, 20, 30, 10, 40]
data_serises = pd.Series(data)
```

```python
mean_val = data_serises.mean()  # 평균값
median_val = data_serises.median()  # 중간값
mode_val = data_serises.mode()[0]  # 최빈값
```

```python
print(mean_val)
print(median_val)
print(mode_val)
```

```python
data = {'점수': [55, 60, 65, 70, 75, 80, 85, 90, 95]}
```

```python
df = pd.DataFrame(data)
```

```python
# 분산
var_val = df['점수'].var()
print(var_val)
```

```python
# 표준편차
std_val = df['점수'].std()
print(std_val)
```

```python
# 사분위 점수
q1_val = df['점수'].quantile(0.25)  # 1사분위수
q2_val = df['점수'].quantile(0.5)   # 2사분위수
q3_val = df['점수'].quantile(0.75)  # 3사분위수
q4_val = df['점수'].max()           # 4사분위
```

```python
print(q1_val, q2_val, q3_val, q4_val)
```

```python
print('IQR', q3_val - q1_val)
```
