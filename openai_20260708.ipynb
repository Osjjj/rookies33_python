OpenAI.API

```python
from dotenv import load_dotenv
import os

load_dotenv()

# 사용할 키 가져오기
OPENAI_API_KEY = os.getenv('OPENAI_API_KEY')
```

```python
from openai import OpenAI
```

```python
client = OpenAI(api_key=OPENAI_API_KEY)
```

```python
# 응답 생성 : Responses
response = client.responses.create(
    model = 'gpt-5.5',
    input='스타워즈 시리즈의 다스베이더의 역할에 대해 한 줄로 설명해줘'
)
```

```python
response
```

```python
print(response.output_text)
```

```python
response = client.responses.create(
    model = 'gpt-5.5',
    instructions='당신은 영화 평론가야!',
    input='스타워즈 시리즈의 다스베이더의 역할에 대해 설명해줘'
)

print(response.output_text)
```

```python
response = client.responses.create(
    model = 'gpt-5.5', 
    input=[
        {
            'role':'developer',
            'content':'음식에 대한 이야기 하는것을 좋아해'
        },
        {
            'role':'user',
            'content':'오늘 저녁 식사는 무엇을 먹을까?'
        }
    ]
)

print(response.output_text)
```

유치원 선생님 앱 => 질문을 받아서 답변하는 앱

```python
user_input = input('안녕하세요! 어린이 여러분!!! 궁금한게 있으면 물어보세요~')
print('질문:', user_input)
```

```python
response = client.responses.create(
    model = 'gpt-5.5', 
    input=[
        {
            'role':'developer',
            'content':'너는 유치원 선생님이야. 아이들을 대하는 것처럼 이야기해줘'
        },
        {
            'role':'user',
            'content': user_input
        }
    ]
)

print(response.output_text)
```
